```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
details = []

for id, pannelli in tqdm(df_dettaglio_pannelli.iterrows(), total=df_dettaglio_pannelli.shape[0]):
  

      
      product_id = pannelli['id']
      link = pannelli["url"]  
      modello = ''
      dimensioni = ''
      range_potenza = ''
      tipologia= ''
      price = ''
      region = ''
      produttore = ''
      peso = ''
      
      

      try:
        wd.set_window_size(1920, 1080)
        wd.get(link)
        try:
          WebDriverWait(wd, 1)
        except Exception as e:
          continue

        #wd.save_screenshot(f'screenshot_{product_id}.png')
        time.sleep(2)
        modello_list = wd.find_elements_by_css_selector("div:nth-child(1) > h1")
        #dimensioni_list = wd.find_elements_by_css_selector("p.reprint")
        range_potenza_list = wd.find_elements_by_css_selector("tr.enf-grey-back > td")
        tipologia_list = wd.find_elements_by_css_selector("div:nth-child(3) > table > tbody > tr:nth-child(1) > td")
        price_list = wd.find_elements_by_css_selector("div.mk-price-body > div > span > b")
        region_list = wd.find_elements_by_css_selector("div:nth-child(3) > table > tbody > tr:nth-child(3) > td")
        produttore_list = wd.find_elements_by_css_selector("div.mk-title > div:nth-child(1) > span")
        #peso_list = wd.find_elements_by_css_selector("#format > div")
        
        

        if(len(modello_list) > 0):
          modello = modello_list[0].text     
        if(len(dimensioni_list) > 0):
          dimensioni = dimensioni_list[0].text
        if(len(range_potenza_list) > 0):
          range_potenza = range_potenza_list[0].text
        if(len(tipologia_list) > 0):
          tipologia = tipologia_list[0].text
        if(len(price_list) > 0):
          price = price_list[0].text     
        if(len(region_list) > 0):
          region = region_list[0].text
        if(len(produttore_list) > 0):
          produttore = produttore_list[0].text        
        if(len(peso_list)>0):
          peso = peso_list[0].text
            
          
      except Exception as e:
        print(e)
      details.append({'url': link,
                      'modello': modello,
                      'range_potenza': range_potenza,
                      'tipologia': tipologia,
                      'price': price,
                      'region': region,
                      'produttore': produttore,
                      'id' : product_id})
  
print('numero righe lette ' + str(len(details)))
pprint.pprint(details)
```
